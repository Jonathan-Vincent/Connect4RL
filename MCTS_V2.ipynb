```python
class node:
    """
    Class for Tree
    name : string
    parent : node
    num : float
    value : float
    variance: float
    children : list
    """
    def __init__(self, name = "", parent=None, num = 0, value = np.inf, variance = np.inf, n = 1):
        self.name = name
        self.parent = parent
        self.num = num
        self.value = value
        self.variance = variance
        self.children = []
        self.n = n
        
    def add_child(self,name = "", num = 0, value = None):
        if value == None:
            value = self.value
        child = node(name = name, parent=self, num = num, value = value)
        self.children.append(child)
        
    def update_distribution(self):
        self.value = min([i.value for i in self.children])
        self.variance = C*np.sqrt(np.log(T)/self.n)
    
    def sample(self):
        for i in self.children:
            if i.children == True:
                i.update_distribution()
        return(self.children[np.argmax([-i.value + i.variance for i in self.children])])
    
    def print_tree(self,depth = 0):
        if self.children == []:
            print("\t"*depth,self)
        else:
            print("\t"*depth,self)
            for i in self.children:
                i.print_tree(depth+1)
    
    def __repr__(self):
        if self.parent == None:
            return "Root(name: %s, num: %s, value: %s, variance: %s, n: %s)"% (self.name, str(self.num), str(self.value), str(self.variance), str(self.n-1))
        else:
            return "Node(name: %s, num: %s, value: %s, variance: %s, n: %s)"% (self.name, str(self.num), str(self.value), str(self.variance), str(self.n-1))

    def __str__(self):
        if self.parent == None:
            return "Root(name: %s, num: %s, value: %s, variance: %s, n: %s)"% (self.name, str(self.num), str(self.value), str(self.variance), str(self.n-1))
        else:
            return "Node(name: %s, num: %s, value: %s, variance: %s, n: %s)"% (self.name, str(self.num), str(self.value), str(self.variance), str(self.n-1))
```

```python
class MCTS:
    """
    Class for the Tree Search algorithm
    It generates the tree step by step while evaluating
    """
    
    def __init__(self, root, c):
        self.root = root
        global C
        C = c
        global T
        T = 1
        
    def generate_sample(self):
        global T
        T+=1
        self.root.n += 1
        node = self.root
        check = True
        while True:
            if node.children == []:
                check = self.generate_children(node)
            if check == False:
                break
            node = node.sample()
            node.n += 1
        node.value = self.evaluate(node)
        node.variance = 0
        while node.parent != None:
            node = node.parent
            node.update_distribution()
            if sum([i.variance for i in node.children])==0:
                node.variance = 0
        self.root = node
    
    def generate_children(self,node):
        children = childs(node)#SQL
        if children == []:
            return(False)
        for i in children:
            node.add_child(i[0],i[1])
        return(True)
    
    def evaluate(self, node):
        #Evaluation function for node
        value = 0
        for i in range(len(node.name)-4):
            value += (node.num) % (i+10)
            node = node.parent
        return(value+0.0000001)
    
    def return_best(self):
        node = self.root
        while True:
            if node.children != []:
                node = node.children[np.argmin([i.value for i in node.children])]
            else:
                break
        print(node)
        return(self.evaluate(node))
    
```

```python
def childs(node):
    if len(node.name)>8:
        return([])
    if node.value != np.inf:
        return([])
    l = random.choices([0,2,3,4],[0.1,0.6,0.2,0.1])[0]
    if node.parent == None: l=4
    childrenlist = []
    for i in range(l):
        childrenlist.append([node.name+str(i), np.random.poisson(100)])
    return(childrenlist)
```
